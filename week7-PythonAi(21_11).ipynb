```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=tf.Variable(3.0)

with tf.GradientTape() as tape:

  y=x**2

  dy_dx=tape.gradient(y,x)

print(f'At x={(x).numpy()}, y is {y}')  #just use numpy so we change tensor number format to numpy formate
print(f'The dy/dx is : {dy_dx.numpy()}')
```

```python
x= tf.Variable(2.0)
y= tf.Variable(3.0)
#use gradienttape to record the operations
with tf.GradientTape() as tape:
  z=x**2+y**3


dz_dx,dz_dy=tape.gradient(z,[x,y]) #pay attention to the syntax, when more variable considered

print("The value of z when x=2 and y=3:", z.numpy())
print("The gradient of z with respect to x:", dz_dx.numpy())
print("the gradient of z with respect to y:", dz_dy.numpy())
```

```python
x_vals = np.linspace(-3,3,30) #range of x
y_vals = np.linspace(-3,3,30) #range of y
x, y = np.meshgrid(x_vals, y_vals)

#compute z for each pair of (x,y)
z = x*2 + y*2

#compute gradient for each point
x_tf = tf.Variable(x, dtype=tf.float32)
y_tf = tf.Variable(y, dtype=tf.float32)

with tf.GradientTape(persistent=True) as tape:
    tape.watch(x_tf)
    tape.watch(y_tf)
    z_tf = x_tf*2 + y_tf*2

dz_dx = tape.gradient(z_tf,x_tf)
dz_dy = tape.gradient(z_tf,y_tf)

dz_dx = dz_dx.numpy()
dz_dy = dz_dy.numpy()

fig = plt.figure(dpi=300)
ax= fig.add_subplot(111, projection='3d')
ax.plot_surface(x,y,z,cmap='plasma',alpha=0.9)

ax.quiver(x,y,z,dz_dx,dz_dy,0,color=[0.1,0.5,0.5],linewidth=0.5,length=0.4,normalize=True)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('3D plot of z = x^2 + y^3 its gradient')
plt.show()
```

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the range of x and y values
x_vals = np.linspace(-3, 3, 30)
y_vals = np.linspace(-3, 3, 30)

# Create a meshgrid for x and y
x, y = np.meshgrid(x_vals, y_vals)

# Convert x and y to TensorFlow variables
x_tf = tf.Variable(x, dtype=tf.float32)
y_tf = tf.Variable(y, dtype=tf.float32)

# Use tf.GradientTape to compute the gradients
with tf.GradientTape(persistent=True) as tape:
    tape.watch(x_tf)
    tape.watch(y_tf)

    # Define the function z = x^2 + y^2
    z_tf = x_tf**2 + y_tf**2

# Compute the gradients of z with respect to x and y
dz_dx = tape.gradient(z_tf, x_tf)
dz_dy = tape.gradient(z_tf, y_tf)

# Convert the gradients to NumPy arrays for plotting
dz_dx = dz_dx.numpy()
dz_dy = dz_dy.numpy()

# Plotting
fig = plt.figure(dpi=300)
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(x, y, z_tf.numpy(), cmap='plasma', alpha=0.8)

# Plot the gradient vectors using quiver
ax.quiver(x, y, z_tf.numpy(), dz_dx, dz_dy, 0, color=[0.1, 0.5, 0.5], linewidth=0.5, length=0.4, normalize=True)

# Set labels for the axes
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```
