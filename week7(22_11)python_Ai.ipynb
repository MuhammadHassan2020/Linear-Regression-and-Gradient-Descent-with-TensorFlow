```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

x_train = np.array([0, 1, 2, 3, 4, 5], dtype=np.float32)
y_train = np.array([0, 2, 4, 6, 8, 10], dtype=np.float32)

w = tf.Variable(0.0, dtype=tf.float32)  # Corrected: Initialize as float32
b = tf.Variable(0.0, dtype=tf.float32)  # Corrected: Initialize as float32

def train(w, b, x_train, y_train, learning_rate=0.01, epochs=200):
    #optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)
    optimizer = tf.keras.optimizers.SGD(learning_rate=learning_rate)
    loss_values = []

    for epoch in range(epochs):
        with tf.GradientTape() as tape:
            y_pred = w * x_train + b
            loss = tf.reduce_mean(tf.square(y_train - y_pred))

        gradients = tape.gradient(loss, [w, b])
        optimizer.apply_gradients(zip(gradients, [w, b]))
        loss_values.append(loss.numpy())

        if epoch % 100 == 0:
            print(f"Epoch {epoch}, loss: {loss.numpy()}, w: {w.numpy()}, b: {b.numpy()}")

    # Corrected: Use the list `loss_values` in the plot, not a string
    plt.plot(loss_values)
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title("Loss over time")
    plt.show()

    return w, b

trained_w, trained_b = train(w, b, x_train, y_train)
print("Trained weight (w):", trained_w.numpy())
print("Trained bias (b):", trained_b.numpy())
```

```python
predictions = [i*trained_w.numpy() + trained_b for i in x_train]
plt.scatter(x_train, y_train , color='blue', label='Data points')
plt.plot(x_train, predictions , color='red', label='Regression Line')
plt.xlabel("X")
plt.xlabel("Y")
plt.legend()
plt.show()
```

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# Training data
X_train_1 = np.array([0, 1, 2, 3, 4, 5], dtype=np.float32)
y_train_1 = np.array([0, 2, 4, 6, 8, 10], dtype=np.float32)

X_train_2 = np.array([0, 1, 2, 3, 4, 5], dtype=np.float32)
y_train_2 = np.array([0, 1.8, 45, 6.2, 8, 10], dtype=np.float32)

X_train, y_train = X_train_1, y_train_1  # Use Dataset 1
# X_train, y_train = X_train_2, y_train_2  # Uncomment to use Dataset 2

# Define the learning rate
lr = 0.01

# Define a dictionary of optimizers
optimizers = {
    'SGD': tf.keras.optimizers.SGD(learning_rate=lr),
    'SGD with Momentum': tf.keras.optimizers.SGD(learning_rate=lr, momentum=0.9),
    'Adam': tf.keras.optimizers.Adam(learning_rate=lr),
    'AdamW': tf.keras.optimizers.AdamW(learning_rate=lr, weight_decay=1e-4),
    'RMSprop': tf.keras.optimizers.RMSprop(learning_rate=lr)
}

# Choose the optimizer by uncommenting one of the following lines:
optimizer = optimizers["SGD"]
# optimizer = optimizers["SGD with Momentum"]
# optimizer = optimizers["Adam"]
# optimizer = optimizers["AdamW"]
# optimizer = optimizers["RMSprop"]



def train(w, b, X_train, y_train, optimizer, epochs=200):
    loss_values = []

    for epoch in range(epochs):
        with tf.GradientTape() as tape:
            # Predict
            y_pred = w * X_train + b
            # Mean squared error loss
            loss = tf.reduce_mean(tf.square(y_train - y_pred))

        # Compute gradients of loss with respect to w and b
        gradients = tape.gradient(loss, [w, b])

        # Apply gradients to update w and b
        optimizer.apply_gradients(zip(gradients, [w, b]))

        # Track loss for plotting
        loss_values.append(loss.numpy())

        # Optional: print progress
        if epoch % 50 == 0:
            print(f'Epoch {epoch}, Loss: {loss.numpy():.4f}, w: {w.numpy():.4f}, b: {b.numpy():.4f}')

    # Plot the loss over epochs
    plt.plot(loss_values)
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title(f'Loss Over Time with {type(optimizer)._name_}')
    plt.show()

    return w, b

# Initialize weight and bias
w = tf.Variable(0.0)
b = tf.Variable(0.0)
# Train the model
trained_w, trained_b = train(w, b, X_train, y_train, optimizer,400)

# Display trained weight and bias
print('Trained weight (w):', trained_w.numpy())
print('Trained bias (b):', trained_b.numpy())

#visualize the trained model

plt.scatter(X_train, y_train, label='Training Data')
plt.plot(X_train, trained_w.numpy()* X_train +trained_b.numpy(), color='red', label='learned line')
plt.xlabel('x_train')
plt.ylabel('y_train')
plt.legend()
plt.show()
```
